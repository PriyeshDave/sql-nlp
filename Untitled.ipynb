```python
import pandas as pd
import numpy as np
data=pd.read_csv('Augmented.csv')
data = data.rename_axis('ID').values
l=data.tolist()
for i in range(len(l)):
    l[i][1]=l[i][1].lower()
    l[i][2]=l[i][2].lower()
```

```python
import re
for i in range(len(l)):
    x=re.findall('(?<=select\s).+?(?=\sfrom)',l[i][2])
    print(x[0])
    l[i].append(x[0])
    y=re.findall('(?<=from\s)[a-z\_]+',l[i][2])
    l[i].append(y[0])
```
